```python
import sys
!{sys.executable} -m pip install pycryptodome

from Crypto.Cipher import AES
from Crypto.Util.Padding import pad, unpad
from Crypto.Random import get_random_bytes
import random
import base64

# === Step 1: Generate random IoT sensor readings ===
temperature = round(random.uniform(20.0, 35.0), 2)
humidity = round(random.uniform(30.0, 70.0), 2)
data = f"Temperature: {temperature}C, Humidity: {humidity}%"
print("[IoT Device] Original Data:", data)

# === Step 2: Encrypt data using AES (CBC mode) ===
key = get_random_bytes(16)
cipher = AES.new(key, AES.MODE_CBC)
ct_bytes = cipher.encrypt(pad(data.encode(), AES.block_size))
iv = base64.b64encode(cipher.iv).decode('utf-8')
ct = base64.b64encode(ct_bytes).decode('utf-8')
print("\n[IoT Device] Encrypted Data Sent to Server:", ct)

# === Step 3: Simulate transmission and decrypt on server side ===
iv_decoded = base64.b64decode(iv)
ct_decoded = base64.b64decode(ct)
cipher_dec = AES.new(key, AES.MODE_CBC, iv_decoded)
pt = unpad(cipher_dec.decrypt(ct_decoded), AES.block_size).decode('utf-8')

# === Step 4: Display results ===
print("\n[Server] Decrypted Data Received:", pt)
```
